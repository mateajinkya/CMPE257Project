1. Load the Processed Dataset



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice theme
sns.set(style='whitegrid')

# Load the cleaned dataset
df = pd.read_csv('student+performance/student/student_combined_processed.csv', sep=',')

print("Data loaded successfully!")
print("Shape:", df.shape)
df.head()
```

2. Summary Statistics and Data Info


```python
# View column data types and non-null counts
df.info()

# Get summary statistics for all numerical columns
df.describe()
```

3. Pass/Fail Distribution

```python
# Plot overall pass/fail distribution
# Force proper comma delimiter
import pandas as pd

file_path = 'student+performance/student/student_combined_processed.csv'
df = pd.read_csv(file_path, sep=',')
print(df.columns.tolist())
sns.countplot(x='pass', data=df)
plt.title('Pass/Fail Distribution')
plt.xlabel('Pass (1) vs Fail (0)')
plt.ylabel('Count')
plt.show()
```

 4. Pass/Fail Distribution by Subject

```python
# Plot pass/fail split for each subject
sns.countplot(x='pass', hue='subject', data=df)
plt.title('Pass/Fail by Subject')
plt.xlabel('Pass (1) vs Fail (0)')
plt.ylabel('Count')
plt.legend(title='Subject (0=Math, 1=Portuguese)')
plt.show()
```

5. Correlation Heatmap

```python
plt.figure(figsize=(16, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

6. Feature Distributions (Histograms)

```python
# Pick a few important or interesting features
key_features = ['studytime', 'failures', 'absences', 'goout', 'G3']

for feature in key_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

 7. Pairplots of Key Features vs Pass/Fail

```python
# Visualize relationships between features colored by pass/fail
sns.pairplot(df[['studytime', 'failures', 'absences', 'G3', 'pass']], hue='pass')
plt.suptitle("Pairplot of Key Features by Pass/Fail", y=1.02)
plt.show()
```

8. Correlation with Targets (Feature Importance Preview)

```python
# Correlation with G3 (regression target)
corr_with_g3 = df.corr()['G3'].sort_values(ascending=False)
print("🔢 Top features positively correlated with G3 (Final Grade):")
print(corr_with_g3.head(10))

# Correlation with pass/fail (classification target)
corr_with_pass = df.corr()['pass'].sort_values(ascending=False)
print("\\n🔢 Top features positively correlated with Pass (1):")
print(corr_with_pass.head(10))
```
